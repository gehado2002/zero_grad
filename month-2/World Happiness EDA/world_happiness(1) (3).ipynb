```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

fig, ax = plt.subplots(2, 3, figsize=(20, 12))

cols = ['gdp', 'social support','life_expectancy',
        'freedom', 'generosity', 'corruption']

for i, col in enumerate(cols):
    sns.regplot(
        x=col, y='happiness', data=df, ax=ax[i//3, i%3],
        scatter_kws={'color': '#ff0000', 'alpha': 0.6},
        line_kws={'color': '#cc0000', 'linewidth': 2}
    )

plt.tight_layout()
plt.savefig("plot.png")   # ← ده اللي يحفظ الصورة
plt.show()

from IPython.display import Image
Image("plot.png")
```

<div style="padding:20px; border-radius:15px; background: linear-gradient(90deg, red, pink); color:white; font-family:Arial; line-height:1.8;">

<p style="font-size:16px; text-align:center; margin-bottom:15px;">
This EDA explores factors affecting <b>happiness</b> across countries. Key variables include <b>GDP 💰</b>, <b>Social Support 🤝</b>, <b>Life Expectancy 🩺</b>, <b>Freedom 🕊️</b>, <b>Generosity 🎁</b>, and <b>Corruption ⚖️</b>.
</p>

<p style="font-size:16px; text-align:center; margin-bottom:15px;">
Results show strong positive correlations with GDP, Social Support, Life Expectancy, and Freedom, moderate effect from Generosity, and negative correlation with Corruption.
</p>

<p style="font-size:16px; text-align:center;">
These insights highlight the most influential factors on happiness globally.
</p>

</div>

```python
df['happiness_level'] = pd.cut(df['happiness'],
                               bins=[0, 5, 10],
                               labels=['Low', 'High'])

fig, ax = plt.subplots(2, 3, figsize=(20, 12))
cols = ['gdp', 'social support', 'life_expectancy',
        'freedom', 'generosity', 'corruption']

red_palette = sns.color_palette("Reds", n_colors=2)

for i, col in enumerate(cols):
    sns.histplot(
        data=df,
        x=col,
        hue='happiness_level',
        multiple='stack',
        bins=20,
        palette=red_palette,
        ax=ax[i//3, i%3]
    )
    ax[i//3, i%3].set_title(f'{col.title()} by Happiness Level')

plt.tight_layout()
plt.savefig("plot.png")   # ← ده اللي يحفظ الصورة
plt.show()

from IPython.display import Image
Image("plot.png")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import warnings

warnings.filterwarnings("ignore", category=FutureWarning)

# --- Ensure only numeric columns are selected ---
numeric_df = df.select_dtypes(include=['number'])

# --- Compute correlations ---
corr = numeric_df.corr()

# --- Plot heatmap in red gradient ---
plt.figure(figsize=(12, 10))
sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    square=True,
    linewidths=0.5,
    cmap="Reds"  # shades of red
)
plt.title('Correlation Heatmap of Numeric Features', fontsize=16, fontweight='bold')
plt.savefig("plot.png")   # ← ده اللي يحفظ الصورة
plt.show()

from IPython.display import Image
Image("plot.png")
```

<h2>Happiness and Its Key Drivers 🌍✨</h2>

<p>Happiness levels across countries are strongly influenced by <strong>GDP per capita</strong> 💰 and <strong>healthy life expectancy</strong> 🏥, both showing a positive correlation.</p>

<p>In contrast, <strong>perceptions of corruption</strong> ⚖️ exhibit a strong negative correlation, indicating that higher corruption is associated with lower happiness.</p>

<p><strong>💡 Insight:</strong> Income and health enhance well-being, while corruption diminishes it.</p>

</div>


```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore", category=FutureWarning)

# Custom red shades
red_palette = ['#ffcccc', '#ff9999', '#ff6666', '#ff3333', '#cc0000', '#990000', '#660000']

# Variables to plot
variables = ['happiness', 'gdp', 'social support', 'life_expectancy',
             'freedom', 'generosity', 'corruption']

# Loop through variables
for var in variables:
    plt.figure(figsize=(12, 6))
    sns.boxplot(x='region', y=var, data=df, palette=red_palette)
    sns.stripplot(x='region', y=var, data=df, color='black', alpha=0.3, jitter=True)
    plt.xticks(rotation=45)
    plt.title(f"{var.replace('_', ' ').title()} Distribution by Region", fontsize=16)
    plt.ylabel(var.replace('_', ' ').title())
    plt.xlabel("Region")
    plt.tight_layout()
plt.savefig("plot.png")   # ← ده اللي يحفظ الصورة
plt.show()

from IPython.display import Image
Image("plot.png")
```

<div style="background: linear-gradient(90deg,#8B0000,#FF6347); padding:14px; border-radius:10px; color:white; font-family:Arial, sans-serif;">
  <h3>📊 Boxplot Summary</h3>
  <ul>
    <li>Higher GDP, social support, and life expectancy link to higher happiness.</li>
    <li>Low corruption also supports higher well-being.</li>
    <li>Generosity shows weaker differences between regions.</li>
  </ul>
</div>


```python
region_happiness = df.groupby('region')['happiness'].mean().sort_values(ascending=False)
top_region = region_happiness.index[0]

df_top = df[df['region'] == top_region]

cols = ['gdp', 'social support', 'life_expectancy', 'freedom', 'generosity', 'corruption']
correlations = df_top[cols + ['happiness']].corr()['happiness'].drop('happiness')
```

```python
region_low_happiness = df.groupby('region')['happiness'].mean().sort_values()
low_region_name = region_low_happiness.index[0]

df_low_region = df[df['region'] == low_region_name]

cols = ['gdp', 'social support', 'life_expectancy', 'freedom', 'generosity', 'corruption']
correlations = df_low_region[cols + ['happiness']].corr()['happiness'].drop('happiness')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore", category=FutureWarning)
fig, axes = plt.subplots(1, 2, figsize=(20,10))
fig.suptitle("Comparison of Factors Affecting Happiness: Highest vs Lowest Regions", fontsize=18, fontweight='bold')

colors_high = ['#ff4d4d' if col != correlations.idxmax() else '#b30000' for col in correlations.index]
sns.barplot(x=correlations.index, y=correlations.values, palette=colors_high, ax=axes[0])
axes[0].set_title(f"Highest Region: {top_region}", fontsize=16)
axes[0].set_ylabel("Correlation with Happiness")
axes[0].tick_params(axis='x', rotation=45)

colors_low = ['#ff4d4d' if col != correlations.idxmin() else '#b30000' for col in correlations.index]
sns.barplot(x=correlations.index, y=correlations.values, palette=colors_low, ax=axes[1])
axes[1].set_title(f"Lowest Region: {low_region_name}", fontsize=16)
axes[1].set_ylabel("Correlation with Happiness")
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.savefig("plot.png")   # ← ده اللي يحفظ الصورة
plt.show()

from IPython.display import Image
Image("plot.png")
```

<div style="background: linear-gradient(90deg, red, white ); padding: 20px; border-radius: 15px; color: white; font-family: Arial, sans-serif;">

# Happiness Insights 🌟

- Highest Region:  → Strongest positive factor: Social Support</span>  
- Lowest Region:  → Strongest negative factor: Corruption</span>

</div>


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

cols1 = ['social support','freedom','corruption']
cols2 = ['gdp', 'life_expectancy']

top_country = df.loc[df['happiness'].idxmax()]
low_country = df.loc[df['happiness'].idxmin()]

fig, axes = plt.subplots(1, 2, figsize=(20, 6))
fig.suptitle(f"Comparison of Factors: {top_country['country']} vs {low_country['country']}", fontsize=16, fontweight='bold')

# ---- Factors Set 1 ----
comparison_df1 = pd.DataFrame({
    'Factor': cols1 * 2,
    'Value': list(top_country[cols1]) + list(low_country[cols1]),
    'Country': [top_country['country']] * len(cols1) + [low_country['country']] * len(cols1)
})
order1 = comparison_df1.groupby('Factor')['Value'].mean().sort_values(ascending=False).index
sns.barplot(
    data=comparison_df1,
    x='Factor', y='Value',
    hue='Country',
    order=order1,
    palette=['#ff4d4d', '#b30000'],
    ax=axes[0]
)
axes[0].set_title("Factors Set 1")
axes[0].tick_params(axis='x', rotation=45)
axes[0].set_ylabel("Value")

# ---- Factors Set 2 ----
comparison_df2 = pd.DataFrame({
    'Factor': cols2 * 2,
    'Value': list(top_country[cols2]) + list(low_country[cols2]),
    'Country': [top_country['country']] * len(cols2) + [low_country['country']] * len(cols2)
})
order2 = comparison_df2.groupby('Factor')['Value'].mean().sort_values(ascending=False).index
sns.barplot(
    data=comparison_df2,
    x='Factor',
    y='Value',
    hue='Country',
    order=order2,
    palette=['#ff4d4d', '#b30000'],
    ax=axes[1]
)
axes[1].set_title("Factors Set 2")
axes[1].tick_params(axis='x', rotation=45)
axes[1].set_ylabel("Value")

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.savefig("plot.png")   # ← ده اللي يحفظ الصورة
plt.show()

from IPython.display import Image
Image("plot.png")
```
